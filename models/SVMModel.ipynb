```python
# print(similarity[3][5])
def mood(mood):
        
    if mood == 'Happy':
        mood = 0
    elif mood == 'Sad':
        mood = 1
    elif mood == 'Calm':
        mood = 2
    elif mood == 'Sleepy':
        mood = 3
    elif mood == 'Energised':
        mood = 4
    elif mood == 'Aroused':
        mood = 5
    elif mood == 'Angry':
        mood = 6
    elif mood == 'Chill':
        mood = 7
    
    return mood
```

```python

def check_similarity(predictions, y_test):
    similarity = [
    [ 1,  0, .5, .3, .5, .7,  0, .6],
    [ 0,  1, .5, .4, .3, .5, .7, .5],
    [.5, .5,  1, .8,  0, .5,  0, .9],
    [.3, .4, .8,  1,  0, .2,  0, .5],
    [.5, .3,  0,  0,  1, .3, .8,  0],
    [.7, .5, .5, .2, .3,  1, .2, .6],
    [ 0, .7,  0,  0, .8, .2,  1,  0],
    [.6, .5, .9, .5,  0, .6,  0,  1]
    ]
    sum_similarity = 0
    test_y = y_test.to_numpy()

    for row in range(len(predictions)):
        model_mood = mood(predictions[row])
        y_mood = mood(test_y[row])
        sum_similarity += similarity[model_mood][y_mood]
    
    accuracy = sum_similarity/len(predictions)
    return accuracy

    
```
